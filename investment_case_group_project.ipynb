### Checkpoint 4: Sector Analysis 1


- #### Extract the primary sector of each category list from the category_list column

```python
# Split 'category_list' by '|', then pick up 1st value as primary sector.
master_frame['primary_sector']=list(master_frame['category_list'].str.split('|').str)[0]
```

```python
# verify the data is correct. 
# eg. Application Platforms will be considered as primary sector for Application Platforms|Real Time|Social Network Media.
master_frame[['category_list','primary_sector']].head()
```

- #### Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that Others is also considered one of the main sectors)

```python
# Read mapping file
import re
sector_mapping_frame=pd.read_csv('mapping.csv', encoding = 'latin-1')
sector_mapping_frame.head()
```

```python
##Removing the NaN values from category list and also do the spelling check for the Analytics values
sector_mapping_frame = sector_mapping_frame[~pd.isnull(sector_mapping_frame['category_list'])]
#sector_list = re.findall("[A-Za-z]{1,}0[A-Za-z]{1,}" , str(sector_mapping_frame['category_list']))
```

```python
sector_mapping_frame['category_list'] = sector_mapping_frame['category_list'].replace("0$",'na', regex = True)
sector_mapping_frame.to_csv("sector_mapping.csv")
sector_mapping_frame
```

```python
# Converting the 0 in values to None for easier removal.
sector_mapping_frame['Automotive & Sports']=sector_mapping_frame['Automotive & Sports'].apply(lambda x: 'Automotive & Sports' if x else None)
```

```python
sector_mapping_frame.head()
```

```python
##Creating a function to create Main_Sector Coulmn to map primary sectors to 8 main sectors
def sectorName(x):
    if x[0]:
        return 'Automotive & Sports'
    elif x[1]:
        return 'Blanks'
    elif x[2]:
        return 'Cleantech / Semiconductors' 
    elif x[3]:
        return 'Entertainment'
    elif x[4]:
        return 'Health'
    elif x[5]:
        return 'Manufacturing'
    elif x[6]:
        return 'News, Search and Messaging'
    elif x[7]:
        return 'Others'
    elif x[8]:
        return 'Social, Finance, Analytics, Advertising'
    else:
        return None
```

```python
sector_mapping_frame['main_sector']=sector_mapping_frame[['Automotive & Sports', 'Blanks', 'Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising']].apply(sectorName, axis=1)
```

```python
sector_mapping_frame[['category_list','main_sector']]
```

```python
sector_frame=sector_mapping_frame[['category_list','main_sector']].rename(index=str, columns={'category_list':'primary_sector'})
```

```python
#Merging master frame and the mapping file to link primary sector in master frame to main sectors in mapping
master_frame=pd.merge(master_frame, sector_frame, how='left', on='primary_sector')
```
